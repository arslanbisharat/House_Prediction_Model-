```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('../data/house_price_regression_dataset.csv')
```

```python
numerical_columns = ['Square_Footage', 'Num_Bedrooms', 'Num_Bathrooms', 'Year_Built', 'Lot_Size', 'Garage_Size', 'Neighborhood_Quality', 'House_Price']

for i in numerical_columns: 
    plt.figure(figsize=(6, 4))
    sns.boxplot(data[i])
    plt.title(f"Box Plot of {i}")
    plt.show()
```

```python
for i in numerical_columns[:-1]:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x = data[i], y=data['House_Price'])
    plt.title('Scatter Plot against House Price')
    plt.xlabel(i)
```
